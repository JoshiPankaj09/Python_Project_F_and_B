# Joining these dataframes to create a consolidated dataframe

```python
merged_df = pd.merge(df1,df2,on= 'City_ID', how= 'inner')
final_df = pd.merge(merged_df, df3, on= 'Respondent_ID', how= 'inner')
pd.set_option('display.max_columns',None)
final_df.head()
```

# Exploratory Data Analysis

```python
Gender_Dis = final_df['Gender'].value_counts()
plt.figure(figsize=(5,3))
x=Gender_Dis.plot(kind='bar',color=['blue','pink','yellow'])
for bars in x.containers:
    x.bar_label(bars)
plt.ylabel('Count')
plt.title('Gender Distribution')
plt.show()
```

The gender distribution above shows that most of the people consuming energy drinks are males

```python
# let's see Gender distribution in price brackets which shows purchasing power and price expectations of genders
plt.figure(figsize=(10,4))
sns.set(style='whitegrid')
sns.countplot(x='Price_range', hue='Gender', data=final_df, palette='Set2')
plt.xlabel('Price_range')
plt.ylabel('Count')
plt.title('Gender Distribution in Price Brackets')
plt.show()
```

The Gender distribution in price range also shows that the majority of orders are placed by Males

```python
# let's see consumer's frequency with respect to age
sns.set(style='whitegrid')
plt.figure(figsize=(8,4))
sns.countplot(x='Consume_frequency',hue='Age', data=final_df,palette='Set1')
plt.xlabel('Consumer_frequency')
plt.ylabel('Count')
plt.title('Consume Frequency w.r.t Age Bracket')
plt.show()
```

The above figure shows the consumer frequency of various age groups, We can see that most of the people are from 19-30 age group, people aged 15-45 consume the most with 19-30 age group being most frequent 

```python
#Distribution of purchase location
purchase_location_count = final_df['Purchase_location'].value_counts()
plt.figure(figsize=(5,5))
plt.pie(purchase_location_count,labels=purchase_location_count.index,autopct='%1.1f%%',startangle=90)
plt.axis('equal')
plt.title('Purchase Location Distribution')
plt.show()
```

The above pie chart shows that Online retailers and Supermarkets are the primary distributors of energy drinks, accounting for the majority of sales in this category

```python
#Marketting channel penetration among different age groups
plt.figure(figsize=(8,4))
sns.countplot(x='Age',hue='Marketing_channels',data=final_df)
plt.title('Proportion of Marketing Channels within Age Groups')
plt.show()
```

In the above figure we can say that among the youth, Online ads and TV commercials are the most impactful marketing channels

```python
#Preferred Ingredient in energy drink
plt.figure(figsize=(8,4))
sns.barplot(y='Ingredients_expected', x=final_df.index, data=final_df, estimator=len,ci=None, palette='viridis',orient='h')
plt.title('Preferred or Expected ingredients among people')
plt.ylabel('Count')
plt.show()
```

The above bar plot shows that Caffeine is the most preferred ingredient in energy drink among people 

```python
#Packaging preference
plt.figure(figsize=(8,5))
plt.pie(final_df['Packaging_preference'].value_counts(), labels=final_df['Packaging_preference'].unique(),autopct='%1.1f%%',startangle=90,
        colors=sns.color_palette('Paired'))
plt.title('Distribution of Packaging Preference')
plt.show()
```

The above pie chart shows that People prefer compact and portable cans and find them convenient

```python
#Most Popullar brands (market leaders)
plt.figure(figsize=(8,4))
sns.countplot(y='Current_brands',data=final_df,order=final_df['Current_brands'].value_counts().index,palette='magma')
plt.title('Most Popular Brands In the Market')
plt.xlabel('Count')
plt.show()
```

From the above horizontal bar plot we observe that Cola-Coka and Bepsi are the most popular brands, it also shows position of our brand CodeX, we can have comparative view of our brand with that of others 

```python
# Pie chart distribution of reasons for choosing a brand
plt.figure(figsize=(10,5))
plt.pie(final_df['Reasons_for_choosing_brands'].value_counts(), labels=final_df['Reasons_for_choosing_brands'].unique(),
        autopct='%1.1f%%',startangle=90,
        colors=sns.color_palette('bright6'))
plt.title('Distribution of Reasons for choosing Brands')
plt.show()
```

The above Pie chart shows a balanced distribution of reasons for choosing brands

```python
# let's drill down to see our brand CodeX in comparison with other brands on every Reason given by public
plt.figure(figsize=(8,4))
highlighted_brand = 'CodeX'
palette = ['orange' if brand == highlighted_brand else 'lightgray' for brand in final_df['Current_brands'].unique()]
sns.countplot(x='Reasons_for_choosing_brands',hue='Current_brands', data=final_df,palette=palette)
plt.title('Reason For Choosing Brands')
plt.xlabel('Reason')
plt.ylabel('Count')
plt.legend(title='Reason for choosing',fontsize= '10', bbox_to_anchor=(1.05,1), loc='upper left')
plt.show()
```

The above plot shows our energy drink brand's position among its competitors, it shows why people prefer other energy drink brands over our energy drink, as we see our brand in comparison with other brands on the basis of 5 different reasons

```python
# Reasons people not buying our drink
no_responses= final_df[final_df['Tried_before']=='No']
plt.figure(figsize=(8,4))
sns.countplot(y='Reasons_preventing_trying',data=no_responses,palette='viridis',
              order=no_responses['Reasons_preventing_trying'].value_counts().index)
plt.title('Reasons for Not Trying Our Drink')
plt.xlabel('Count')
plt.ylabel('Reasons')
plt.show()
```

The above plot shows that most of the people say health concern and lack of interest in energy drinks as the primary reasons for not consuming the energy drink

```python
# Brand perception of our brand
# Taste experience also
plt.figure(figsize=(8,4))
selected_brand = 'CodeX'
selected_brand_df = final_df[final_df['Current_brands']==selected_brand]
sns.countplot(x='Brand_perception',data=selected_brand_df, palette='deep')
plt.title(f'Brand Perception for {selected_brand}')
plt.xlabel('Brand Perception')
plt.ylabel('Count')
plt.show()
```

The above bar plot shows that most of the people have neutral perception for our brand

```python
yes_responses= final_df[final_df['Tried_before']=='Yes']
average_rating = yes_responses['Taste_experience'].mean()
print(f'The average rating for CodeX is: {average_rating}')
plt.figure(figsize=(8,4))
sns.countplot(x='Taste_experience',data=yes_responses,palette='deep')
plt.title('Taste Experience for CodeX')
plt.xlabel('Taste Experience Rating')
plt.ylabel('Count')
plt.show()
```

The above bar plot shows the taste experience rating 1 to 5, for our energy drink CodeX. Most of the people rate it 3, overall the average taste experience is 3.26

```python
# let's see how much brand penetration we have across various cities
plt.figure(figsize=(7,4))
selected_brand = 'CodeX'
selected_brand_df = final_df[final_df['Current_brands']==selected_brand]
sns.countplot(y='City',data=selected_brand_df,palette='magma', order=selected_brand_df['City'].value_counts().index)
plt.title('Brand Penetration of Our Brand CodeX')
plt.xlabel('Count')
plt.ylabel('City')
plt.show()
```

From the above bar plot we observe Lucknow, Jaipur, Delhi, Ahmedabad and Kolkata are the cities where brand penetration seems comparatively low

```python
# let's check typical consumption situation of people
plt.figure(figsize=(10,5))
plt.pie(final_df['Typical_consumption_situations'].value_counts(), labels=final_df['Typical_consumption_situations'].unique(),
        autopct='%1.1f%%',startangle=90,
        colors=sns.color_palette('bright6'))
plt.title('Distribution of Typical Consumtion Situations')
plt.show()
```

The above pie chart shows that sports/exercise and Social outings are the situation where people most likely consume these drinks

```python
# Let's explore the reasons why people consume energy drinks
plt.figure(figsize=(10,5))
plt.pie(final_df['Consume_reason'].value_counts(), labels=final_df['Consume_reason'].unique(),
        autopct='%1.1f%%',startangle=90,
        colors=sns.color_palette('muted6'))
plt.title('Reasons For Consuming Energy Drinks')
plt.show()
```

The pie chart shows most of the people consume energy drinks for energy, focus and to combat fatigue

```python
#Let's see what people think about Limited edition packaging, Whether it really impacts their decision of buying drinks or not
plt.figure(figsize=(8,4))
plt.pie(final_df['Limited_edition_packaging'].value_counts(), labels=final_df['Limited_edition_packaging'].unique(),
        autopct='%1.1f%%',startangle=90,
        colors=sns.color_palette('pastel'))
plt.title('Limited Edition Packaging Likelihood')
plt.show()
```

In the above pie chart we can see that while around 40% of the people found limited edition packaging impactful, an equivalent percentage did not perceive any substantial impact, reflecting a balanced distribution of responses

```python
# let's check the price range which people expect for these energy drinks
plt.figure(figsize=(8,4))
plt.pie(final_df['Price_range'].value_counts(), labels=final_df['Price_range'].unique(),
        autopct='%1.1f%%',startangle=90,
        colors=sns.color_palette('Set1'))
plt.title('Price Range Expectations')
plt.show()
```

The above pie chart shows most of the people (around 75%) are capable of spending more than 50 but less than 150 on these energy drinks

```python
# People viewpoint for natural, organic ingredients in the energy drink
plt.figure(figsize=(8,4))
plt.pie(final_df['Interest_in_natural_or_organic'].value_counts(), labels=final_df['Interest_in_natural_or_organic'].unique(),
        autopct='%1.1f%%',startangle=90,
        colors=sns.color_palette('Set1'))
plt.title('Natural Ingredients Likelihood')
plt.show()
```

The pie chart shows almost half of the people look for natural or organic ingredients in their energy drinks

```python
# Now let's see natural ingredient requirement breakdwon in age groups
plt.figure(figsize=(8,5))
sns.histplot(x='Age',hue='Interest_in_natural_or_organic',multiple='stack',data=final_df,palette='Set1')
plt.title('Distribution of people with natural ingredients across Age groups')
plt.xlabel('Age Groups')
plt.ylabel('Count')
plt.show()
```

In the above stacked bar plot also we observe that almost 50% in every age brakcet show interest in natural or organic ingredients in their energy drinks

```python
# What improvements people want from energy drinks making companies in the product
plt.figure(figsize=(10,5))
plt.pie(final_df['Improvements_desired'].value_counts(), labels=final_df['Improvements_desired'].unique(),
        autopct='%1.1f%%',startangle=90,
        colors=sns.color_palette('colorblind'))
plt.title('The Improments People Desire In Energy Drinks')
plt.show()
```

The pie chart shows that about one-third people want more natural ingredients whereas one-fourth would like reduced sugar content

# Conclusion

The findings reveal that males are predominant consumers of energy drinks, and the age group of 19-31 years exhibits the highest frequency of consumption. Supermarkets and online retailers emerge as the preferred purchase points for these beverages, highlighting the convenience and accessibility factors.

Furthermore, our analysis indicates that online ads and TV commercials play a significant role in influencing the youth, making them effective marketing channels. Preferences for caffeine and vitamin-based drinks, along with a preference for canned packaging, underline the importance of portability and health-conscious choices among consumers.

The average rating of 3.26 for our energy drink suggests a moderately positive reception, encouraging us to further enhance and refine our product. Additionally, cities like Lucknow, Jaipur, Delhi, Ahmedabad, and Kolkata show low brand penetration, suggesting potential areas for targeted marketing and expansion.

Understanding that sports, exercise, and social outings constitute the typical consumption situations for 75% of individuals, we can tailor marketing strategies to align with these scenarios. Moreover, the majority of consumers expect energy drink prices to fall between 50 to 150 Rupees, indicating a price sensitivity that should be considered in pricing strategies.

Lastly, the notable preference for natural ingredients among 50% of consumers signals a growing demand for healthier formulations. This insight can guide product development efforts to cater to the evolving preferences of the market. In conclusion, our Python-based data analysis provides a solid foundation for informed decision-making, enabling us to optimize marketing strategies, product formulations, and market expansion plans.
